```python
import pandas as pd
from diversity_sampling.models import TrainingDynamics, CoreSetSelector
```

```python
train_set = pd.read_csv("diversity_sampling/dataset/real_set.csv")
```

```python
selector = CoreSetSelector()

finetune_logits = selector.get_finetune_logits(
    train_set=train_set,
    batch_size=16,
    num_epochs=3,
    lr=5e-4
)

all_training_dynamics: list[TrainingDynamics] = finetune_logits
```

```python
scores = selector.evaluate_data_contribution(training_dynamics=all_training_dynamics)
data_splits = selector.split_data(dataset=train_set, scores_mapping=scores)
```

# Local Storage

```python
data_splits['augment'].to_csv("diversity_sampling/coresets/augment.csv")
data_splits['retain'].to_csv("diversity_sampling/coresets/retain.csv")
data_splits['prune'].to_csv("diversity_sampling/coresets/prune.csv")
```
